```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as s
```

```python
data_frame = pd.read_csv('Income.csv')
```

```python
data_frame.head()
```

```python
l = data_frame.columns.to_list()
l=l[1:]
for i in l:
    data_frame[i] = pd.to_numeric(data_frame[i], errors='coerce')
```

```python
data_frame['Occupation'] = data_frame['Occupation'].str.lower()
```

```python
data_frame.dropna(subset=['All_weekly'],inplace=True)
```

```python
#data_frame['All_weekly'] = (data_frame['M_weekly'] + data_frame['F_weekly'])/2
data_frame['M_weekly'].fillna(data_frame['All_weekly'] * (2 - data_frame['F_weekly']),inplace=True)
data_frame['F_weekly'].fillna(data_frame['All_weekly'] * (2 - data_frame['M_weekly']),inplace=True)
```

```python
data_frame.dropna(inplace=True)
```

```python
data_frame_1 = data_frame.copy()
data_frame_1['Male_ratio'] = (data_frame_1['M_workers']/data_frame_1['All_workers'])
data_frame_1['Female_ratio'] = (data_frame_1['F_workers']/data_frame_1['All_workers'])
data_frame_1['Male_wage_ratio'] = (data_frame_1['M_weekly']/data_frame_1['All_weekly'])
data_frame_1['Female_wage_ratio'] = (data_frame_1['F_weekly']/data_frame_1['All_weekly'])
data_frame_1
```

```python
data_1a = data_frame_1[data_frame_1['Female_ratio']>0.5]
print(data_1a.head())
data_1a = len(data_1a.index)
data_1a
```

```python
data_1b = data_1a[data_1a['Female_wage_ratio']<0.5]
data_1b = len(data_1b.index)
data_1b
```

```python
data_frame_5 = data_frame_4.copy()
```

```python
X, y = data_frame_5.iloc[:, :-1], data_frame_5.iloc[:, -1]
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
d = []
for i in range(1,21):
    regr = RandomForestRegressor(n_estimators=i, random_state=123)
    regr.fit(X,y)
    y_pred = regr.predict(X)
    r2 = r2_score(y, y_pred)
    d.append(r2)
```

```python
from pandas import DataFrame
data_5 = DataFrame (d,columns=['Score'])
data_5.index = np.arange(1, len(data_5)+1)
```
